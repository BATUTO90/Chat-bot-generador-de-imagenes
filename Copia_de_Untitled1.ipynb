```python
# Primero, importamos lo que vamos a necesitar
from diffusers import StableDiffusionPipeline
import torch

# Verificamos si hay una GPU disponible y la usamos, si no, pues el CPU rifa
device = "cuda" if torch.cuda.is_available() else "cpu"

# Aquí es donde cargas el modelo pre-entrenado de Stable Diffusion.
# Puedes buscar otros modelos en Hugging Face Hub si quieres variar.
# Este es uno de los más comunes para empezar: "runwayml/stable-diffusion-v1-5"
# Si es la primera vez, se va a tardar un poquito en descargar el modelo.
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
pipe = pipe.to(device)

# Ahora, ¡la magia! Escribe lo que quieres generar.
# Este es tu "prompt". Sé descriptivo, ¡échale coco!
prompt = "Un ajolote futurista con luces de neón en una ciudad cyberpunk, estilo acuarela"

# ¡A generar la imagen!
# image contendrá la imagen generada.
image = pipe(prompt).images[0]

# Y ahora, si quieres ver tu obra de arte en el Colab:
from PIL import Image
image.save("mi_imagen_generada.png") # Esto la guarda como un archivo

# Para mostrarla directamente en la salida de la celda:
display(image)

print("¡Órale! Imagen generada y guardada como mi_imagen_generada.png")
```
